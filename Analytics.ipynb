```python
from bs4 import BeautifulSoup
import pandas as pd
import requests
```

```python
url="https://www.analyticsvidhya.com/blog/2023/04/data-science-companies/"
page=requests.get(url)
soup=BeautifulSoup(page.text,'html')
```

```python
table=soup.find_all('table')[0]
```

```python
column_data=table.find('tr')

for head in column_data:
  row=column_data.find_all('td')
  head=[data.text.strip() for data in row]

head
```

```python
df=pd.DataFrame(columns=[head])
df
```

```python
column_data=table.find_all('tr')
for column in column_data[1:]:
  row=column.find_all('td')
  lst=[data.text.strip() for data in row]
  length=len(df)
  df.loc[length]=lst
```

```python
df.to_csv("/content/companies.csv",index=False)
```
